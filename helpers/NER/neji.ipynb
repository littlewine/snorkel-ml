# turn json into standoff format

```python
import json, codecs
```

```python
verbose=True
```

```python
for split_type in split_types:
    input_path = glob.glob(os.path.join(output_path, split_type, 'json','*.json'))
    save_path = os.path.join(output_path, split_type)
    
    for filename in input_path: #for each doc in collection
        doc_id = filename.split("/")[-1].rstrip('.json')

        # parse jsons
        with open(filename, 'rb') as f:
            js = json.load(f)

        i=0
        save_path_i = os.path.join(save_path,"%s.ann"%doc_id)
        with codecs.open(save_path_i, 'wb', "utf-8") as f:
            for sentence in js:
                for term in sentence['terms']:
            #             print term['text'],term['ids'],term['start'],term['end']
                    f.write("T%i\t%s %i %i\t%s\n"%(i+1,"Gene",term['start'],term['end'],term['text']))
                    f.write("N%i\tReference T%i %s\t%s\n"%(i+1,i+1,term['ids'],term['text']))
                    i+=1

            # Write out
            if verbose:
                print "Writing %i entities in: %s"%(i , save_path_i)

```
