```python
def createCAR(colname, outf):
    for i in range(row):
        #print(i)
        PUI = df.ix[i, 'PUI']
        of=open(outf+"PUI"+str(PUI)+".XML","w")
        with open(CARsample) as f:
            line=format(f.readline())
            of.write(str(line))
            while line:
                #print("Line {}: {}".format(cnt, line.strip()))
                line = f.readline()
                matchPUI = re.search("PUI",line)
                matchTitle = re.search("titletext",line)
                matchAbs = re.search("ce:para",line)
                authorKey= re.search("author-keyword",line)
            
                if matchPUI:
                    line=re.sub(r"\>.*\<",">"+str(PUI)+"<",line)
                elif matchAbs:
                    line=re.sub(r"\>.*\<",">"+format(df.iloc[i][colname])+"<",line)
                    #print(line)
                elif matchTitle or authorKey:
                    line=re.sub(r"\>.*\<",">the<",line)
            
                of.write(str(line))
                #print(content)
        of.close()
```
