```python
#put it in a function
def LeadMine_ent_parse(leadmine_output_path):
    entities = pd.read_table(leadmine_output_path)
    entities = entities.rename(columns = { 
        'BegIndex': 'start',
        'EndIndex': 'stop',
        'EntityText': 'name',
        'DocName': 'doc_id'
    })
    # TODO: process only some specific types of LeadMines extracted ones? (eg. drop 'G': generic , eg. sugar)
    entities['type'] = 'Chemical'
    entities['type_start_stop'] = map(lambda x,y,z: "%s %i %i"%(x,y,z) , entities.type, entities.start, entities.stop)
    entities['idx'] = map(lambda x: "T%i"%x, entities.index)
    return entities
```

```python
for split_type in split_types:
    leadmine_output_path = '/home/antonis/data/biocreative6/leadmine/%s_entities.tsv'%split_type
    print 'Parsing entities from', leadmine_output_path
    entities = LeadMine_ent_parse(leadmine_output_path)
    
    ann_save_path = '/home/antonis/data/biocreative6/leadmine/%s/'%split_type
    
    print 'Writing to',ann_save_path
    write_ann_file(entities,save_path= ann_save_path)
    
    print
```
