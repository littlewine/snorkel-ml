```python
from collections import  defaultdict
import glob
```

```python
ann_files = glob.glob("/home/antonis/data/'patents/*/'*/*.ann")
print('Found %i ann files'%len(ann_files))
```

```python
brat_stats = dict()

# filepath = "/home/antonis/data/patents/EP/EP__11_Training_relevancy_relations/EP2266989B1_description_009.ann"

for filepath in ann_files:
    rels = defaultdict(int)
    entts = defaultdict(int)
    with open(filepath) as brat_file:
        for line in brat_file.readlines():
            #get entity type
            if line.startswith("T"):
                spltd = line.split("\t")
    #             print(spltd[0],spltd[1].split()[0])
                entts[spltd[1].split()[0]] +=1

            #count rels
            if line.startswith("R"):
                spltd = line.split("\t")
    #             print(spltd)
                rel, subj, obj = spltd[1].split()
                rels[rel] +=1

        #save them in dict
        brat_stats[filepath.split('/')[-1]] = dict()
        brat_stats[filepath.split('/')[-1]]['relationships'] = rels
        brat_stats[filepath.split('/')[-1]]['entities'] = entts
```

```python
from pprint import pprint
pprint(brat_stats)
```

### find number of rels in all files:

```python
relationships = defaultdict(int)
entities = defaultdict(int)
for f in brat_stats.keys():
    for rel in brat_stats[f]['relationships']:
        relationships[rel] +=1
        
    for ent in brat_stats[f]['entities']:
        entities[ent] +=1
```
